```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)
x_eval = np.linspace(-5,5,100)
plt.plot(x_eval,f(x_eval),'r-',label="Funzione obiettivo")

x_int = np.array([-5,-2,0,2,5])
plt.plot(x_int,f(x_int),'ko',label="Nodi di interpolazione")

plt.title("Funzione e nodi di interpolazione")
plt.legend()
plt.show()
```

$$
P(x)=\sum_{i=0}^{n} f\left(a_{i}\right) \prod_{j \neq i, j=0}^{n} \frac{x-a_{j}}{a_{i}-a_{j}}
$$

```python
def polLagrange(x_int, y_int, x_eval):
  #(x_int,y_int) : nodi ai quali abbiamo informazioni sulla funzione (nodi di interpolazione)
  #x_eval : punti sui quali vogliamo sapere il valore del polinomio interpolante

  y_eval = np.zeros_like(x_eval)
  n = len(x_int) #quanti punti di interpolazione abbiamo

  for i in range(n):
    P = 1
    for j in range(n):
      if j!=i:
        P *= (x_eval-x_int[j]) / (x_int[i]-x_int[j])
    y_eval = y_eval + y_int[i] * P
  
  return y_eval
```

```python
f = lambda x: np.sin(x)
x_eval = np.linspace(-5,5,100)

fig = plt.figure(dpi=600)

plt.plot(x_eval,f(x_eval),'r-',label="Funzione obiettivo")

x_int = np.linspace(-5,5,7)
y_int = f(x_int)
plt.plot(x_int,y_int,'ko',label="Nodi di interpolazione")

y_eval = polLagrange(x_int,y_int,x_eval)
plt.plot(x_eval,y_eval,'g-',label="Polinomio interpolante")

plt.title("Funzione e interpolante")
plt.legend()
plt.show()
```
